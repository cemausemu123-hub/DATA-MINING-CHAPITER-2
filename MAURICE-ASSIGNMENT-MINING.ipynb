```python
from ISLP.models import ModelSpec as MS, summarize, poly
from functools import partial
import statsmodels.api as sm
import pandas as pd
import numpy as np
from sklearn.base import clone

# Load data
Auto = pd.read_csv('C:/Users/DIGITAL AXIS/Desktop/Auto.csv')
Auto['horsepower'] = pd.to_numeric(Auto['horsepower'], errors='coerce')
Auto = Auto.dropna()

# Define boot_OLS function
def boot_OLS(model_spec, response, D, idx):
    D_ = D.iloc[idx]
    Y_ = D_[response]
    X_ = clone(model_spec).fit_transform(D_)
    return sm.OLS(Y_, X_).fit()

# Define boot_SE function
def boot_SE(func, D, n=None, B=1000, seed=0):
    if n is None:
        n = len(D)
    rng = np.random.default_rng(seed)
    results = np.array([
        func(D, rng.choice(n, n, replace=True)).params
        for _ in range(B)
    ])
    return results.std(axis=0, ddof=1)

# Create quadratic model - poly is imported from ISLP.models
quad_model = MS([poly('horsepower', 2, raw=True)])
quad_func = partial(boot_OLS, quad_model, 'mpg')

# Bootstrap
quad_se = boot_SE(quad_func, Auto, B=1000)
print("Bootstrap SE for quadratic model:")
print(quad_se)

```
